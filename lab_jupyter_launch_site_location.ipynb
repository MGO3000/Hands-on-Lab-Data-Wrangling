Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'  # Éxito
    else:
        return 'red'    # Fracaso

spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
marker_cluster = MarkerCluster().add_to(site_map)

# 2. Iterate through spacex_df and add markers for each launch
for index, record in spacex_df.iterrows():
    # Get the launch location
    launch_location = [record["Lat"], record["Long"]]
    
    # Create a marker with a color based on success (green) or failure (red)
    marker = folium.Marker(
        location=launch_location,
        icon=folium.Icon(color='white', icon_color=record['marker_color'], icon="info-sign"),
        popup=f"Launch Site: {record['Launch Site']}<br>Success: {record['class']}"
    )

    # Add the marker to the cluster
    marker_cluster.add_child(marker)

# 3. Display the map with launch markers
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
# Example coastline coordinates (you can refine these based on satellite maps)
coastline_coords = {
    "CCAFS LC-40": (28.56367, -80.57163),
    "CCAFS SLC-40": (28.56367, -80.57163),
    "KSC LC-39A": (28.57205, -80.58528),
    "VAFB SLC-4E": (34.63667, -120.6243)
}
# Create a new column to store distances
launch_sites_df["Distance to Coastline (km)"] = launch_sites_df.apply(
    lambda row: calculate_distance(row["Lat"], row["Long"], 
                                   coastline_coords[row["Launch Site"]][0], 
                                   coastline_coords[row["Launch Site"]][1]), 
    axis=1
)

```

```python
# 1. Loop through each launch site and add a marker for the closest coastline
for index, row in launch_sites_df.iterrows():
    launch_site = row["Launch Site"]
    launch_location = [row["Lat"], row["Long"]]
    
    # Get the closest coastline coordinates
    coastline_location = coastline_coords[launch_site]
    
    # Calculate the distance
    distance = calculate_distance(launch_location[0], launch_location[1], 
                                  coastline_location[0], coastline_location[1])

    # 2. Add a marker for the closest coastline
    folium.Marker(
        location=coastline_location,
        popup=f"Closest Coastline to {launch_site}",
        icon=folium.Icon(color="blue", icon="cloud"),
    ).add_to(site_map)

    # 3. Display distance using a text label marker
    distance_marker = folium.Marker(
        location=[(launch_location[0] + coastline_location[0]) / 2,  # Midpoint
                  (launch_location[1] + coastline_location[1]) / 2],  
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance),
        )
    )
    
    # Add the distance label to the map
    site_map.add_child(distance_marker)

# 4. Display the final map with coastline markers and distances
site_map
```

```python
# 1. Initialize the map
site_map = folium.Map(location=[28.5, -80.6], zoom_start=5)

# 2. Create a marker cluster for launches
marker_cluster = MarkerCluster().add_to(site_map)

# 3. Loop through `spacex_df` and add launch markers (successful=green, failed=red)
for index, record in spacex_df.iterrows():
    launch_location = [record["Lat"], record["Long"]]

    # Create a marker with success (green) or failure (red) based on 'marker_color'
    marker = folium.Marker(
        location=launch_location,
        icon=folium.Icon(color='white', icon_color=record['marker_color'], icon="info-sign"),
        popup=f"Launch Site: {record['Launch Site']}<br>Success: {record['class']}"
    )

    # Add the marker to the cluster
    marker_cluster.add_child(marker)

# 4. Draw PolyLines from launch sites to coastline & add distance labels
for index, row in launch_sites_df.iterrows():
    launch_site = row["Launch Site"]
    launch_location = [row["Lat"], row["Long"]]
    
    # Get the closest coastline coordinates
    coastline_location = coastline_coords[launch_site]
    
    # Calculate the distance
    distance = calculate_distance(launch_location[0], launch_location[1], 
                                  coastline_location[0], coastline_location[1])

    # Add a marker for the closest coastline
    folium.Marker(
        location=coastline_location,
        popup=f"Closest Coastline to {launch_site}",
        icon=folium.Icon(color="blue", icon="cloud"),
    ).add_to(site_map)

    # Add distance label at the midpoint between launch site & coastline
    distance_marker = folium.Marker(
        location=[(launch_location[0] + coastline_location[0]) / 2,  # Midpoint
                  (launch_location[1] + coastline_location[1]) / 2],  
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance),
        )
    )
    site_map.add_child(distance_marker)

    # Draw a PolyLine from launch site to coastline
    line = folium.PolyLine(
        locations=[launch_location, coastline_location],
        weight=2,
        color='blue'
    )
    site_map.add_child(line)

# 5. Display the final map with launches & coastline distances
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site

# Example coordinates for nearest city, railway, and highway (You can refine these with real data)
infrastructure_coords = {
    "CCAFS LC-40": {
        "City": (28.3922, -80.6077),  # Example: Cape Canaveral City
        "Railway": (28.5723, -80.5857),
        "Highway": (28.5638, -80.5708)
    },
    "CCAFS SLC-40": {
        "City": (28.3922, -80.6077),
        "Railway": (28.5723, -80.5857),
        "Highway": (28.5638, -80.5708)
    },
    "KSC LC-39A": {
        "City": (28.3922, -80.6077),
        "Railway": (28.5733, -80.5987),
        "Highway": (28.5721, -80.5774)
    },
    "VAFB SLC-4E": {
        "City": (34.6391, -120.4579),  # Example: Lompoc, CA
        "Railway": (34.6400, -120.4650),
        "Highway": (34.6357, -120.4729)
    }
}
# 1. Loop through each launch site and add distance markers
for index, row in launch_sites_df.iterrows():
    launch_site = row["Launch Site"]
    launch_location = [row["Lat"], row["Long"]]

    # Iterate over each type of infrastructure (City, Railway, Highway)
    for infra_type, infra_location in infrastructure_coords[launch_site].items():
        # Compute distance
        distance = calculate_distance(launch_location[0], launch_location[1], 
                                      infra_location[0], infra_location[1])

        # Add a marker for the infrastructure
        folium.Marker(
            location=infra_location,
            popup=f"Closest {infra_type} to {launch_site}",
            icon=folium.Icon(color="purple" if infra_type == "City" else 
                             "orange" if infra_type == "Railway" else "black"),
        ).add_to(site_map)

        # Add distance label at the midpoint
        distance_marker = folium.Marker(
            location=[(launch_location[0] + infra_location[0]) / 2,  # Midpoint
                      (launch_location[1] + infra_location[1]) / 2],  
            icon=DivIcon(
                icon_size=(20,20),
                icon_anchor=(0,0),
                html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance),
            )
        )
        site_map.add_child(distance_marker)

        # Draw a PolyLine from launch site to infrastructure
        line = folium.PolyLine(
            locations=[launch_location, infra_location],
            weight=2,
            color="purple" if infra_type == "City" else 
                  "orange" if infra_type == "Railway" else "black"
        )
        site_map.add_child(line)

# 2. Display the updated map
site_map
```
